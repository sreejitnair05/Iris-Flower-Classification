```python
feature_columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
X = iris[feature_columns].values
y = iris['species'].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

#Step 3 â€“ Model training.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

#Step 4 â€“ Model Evaluation.

```python
prediction = knn.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,prediction)*100
```

```python
print('Accuracy of our model is  ' + str(round(acc, 2)) + ' %.')
```

#Step 5 â€“ Testing the model.

```python
features = np.array([[6.7, 3.0, 5.2, 2.3]])
# using inputs to predict the output
prediction = knn.predict(features)
```

```python
print("Prediction: {}".format(prediction)) #0=>Iris-setosa , 1=>Iris-versicolor ,2=>Iris-virginica
```
